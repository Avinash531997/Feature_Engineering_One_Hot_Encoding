```python
#Performing one hot encoding and examining number of columns obtained after one hot encoding of these variables
pd.get_dummies(data,drop_first=True).shape
```

It can be observed that when we apply one hot encoding to the categorical features of our dataset,
number of columns have now increased which might lead to curse of dimensionality problem.
So,we come up with a solution according to which for each feature we will not encode all it's categories instead we will encode the top 10 category of each feature.




```python
#Finding 10 most frequent features for each features.
#let's consider X2 first
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
#Let's make a list with most freqent categories of the variable
top_10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)
data[['X2']+top_10].head(20)
```

```python
# get whole set of dummy variables, for all the categorical variables
def one_hot_encoding_top_x(data, variable, top_x_labels):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        data[variable+'_'+label] = np.where(data[variable]==label, 1, 0)
```

```python
data=pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])
#encode X2 into the 10 most frequent  categories
one_hot_encoding_top_x(data,'X2',top_10)
data.head()
```

```python
#Let's make a list with most freqent categories of the variable
top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]
top_10
#encode X1 into the 10 most frequent  categories
one_hot_encoding_top_x(data,'X1',top_10)
data.head()
```

```python
#Let's make a list with most freqent categories of the variable
top_10=[x for x in data.X3.value_counts().sort_values(ascending=False).head(10).index]
top_10
#encode X3 into the 10 most frequent  categories
one_hot_encoding_top_x(data,'X3',top_10)
data.head()
```
